# **1**

```python
def car_dt_frame(x):
  df_vendas = pd.read_csv(x)
  print("O número de linhas do DataFrame é:", df_vendas.shape[0],"\n")
  print("O número de colunas do DataFrame é:", df_vendas.shape[1],"\n")
  print("Os tipos de dados do DataFrame são:","\n","\n", df_vendas.dtypes.value_counts(),"\n")
  print("O nome das colunas e linhas são:", "\n","\n", df_vendas.axes,"\n","\n")
  print("As primeiras 10 linhas são:")
  display(df_vendas.head(10))
  print("\n","As últimas 10 linhas são:")
  display(df_vendas.tail(10))
```
